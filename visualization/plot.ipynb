```python
import json
plt.figure(figsize=(8, 6))
log_path = 'logs/mdanalysis_logs'
colors = ['blue', 'red', 'green', 'purple'] # 红橙黄绿青蓝紫
for i, model in enumerate(['egnn', 'stag_egnn', 'gmn', 'stag_gmn']):
    exp_path = log_path + f'/exp_{model}'
    with open(f"{exp_path}/loss.json") as f:
        loss=json.load(f)

    if model.startswith('stag'):
        model = 'E'+model
    model = model.upper()

    #plt.plot(loss['train loss'],label='Train')
    plt.plot(loss['epochs'],[np.mean(loss['train loss'][i*5:(i+1)*5]) for i in range(len(loss['train loss'])//5)],label=f'{model}(Train)', c=colors[i], linestyle='-')
    plt.plot(loss['epochs'],loss['test loss'],label=f'{model}(Test)', c=colors[i], linestyle='-.')


plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.title("Loss")
plt.savefig('figures/loss_protein.pdf')
```
